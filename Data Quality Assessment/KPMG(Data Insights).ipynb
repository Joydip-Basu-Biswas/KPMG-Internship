```python
from sklearn.preprocessing import LabelEncoder
lb =  LabelEncoder()
df.iloc[:,13] = lb.fit_transform(df.iloc[:,13])
df.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer = imputer.fit(df[['Age']])
df[['Age']] = imputer.transform(df[['Age']])
```

```python
x = df[["Age","state","gender","past_3_years_bike_related_purchases"]]
y = df["owns_car"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.20,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
def models(x_train, y_train):
    
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state = 0)
    log.fit(x_train,y_train)
    
    # Decision tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = "entropy",random_state = 0)
    tree.fit(x_train, y_train)
    
    # Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion = "entropy", random_state = 0)
    forest.fit(x_train, y_train)
    
    # Naive_Bayas
    from sklearn.naive_bayes import GaussianNB
    nb = GaussianNB()
    nb.fit(x_train,y_train)
    
    # svm
    from sklearn.svm import SVC
    svm = SVC(kernel = "linear", random_state = 0)
    svm.fit(x_train,y_train)
    
    # print the model accuracy of training data
    print("[0] Logistic Regression Training Accuricy : ",log.score(x_train, y_train))
    print("[1] decision Tree Classifier Training Accuracy : ",tree.score(x_train, y_train))
    print("[2] Random Forest Classifier Training Accuracy : ", forest.score(x_train, y_train))
    print("[3] Naive Bayas Classifier Training Accuracy : ", nb.score(x_train, y_train))
    print("[4] SVM Training Accuracy : ", svm.score(x_train, y_train))
    
    return log, tree, forest, nb, svm



```

```python
model = models(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print("Model",i)
    cm = confusion_matrix(y_test, model[i].predict(x_test))

    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]
    print(cm)
    print("Testing Accuracy : ", (TP + TN)/(TP + TN + FP + FN))
    print()
```
